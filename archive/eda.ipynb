```python
import requests
import logging as log
import time
import os
from datetime import datetime
import pandas as pd
```

```python

def _getJason(url, f='json'):
    """Internal Function to hit
    """
    log.info('Retrieving Jason for %s' % url)
    data = None
    try:
        resp = requests.get(url)#, data=json.dumps(r_payload),
                            #headers=self.headers)
        if resp.status_code == 200:
            if f=='json':
                data = resp.json()
            else:
                data = resp.text
        else:
            log.error('Unable to get data, status code %s' % resp.status_code)
    except Exception as e:
        log.error("Error getting json %s" % e)

    return data

```

## Coindesk Data

```python
base_url = 'https://api.coindesk.com/charts/data?data=close&output=json'
data_url = '&startdate={0}&enddate={1}&exchanges=bpi&dev=1&index={2}'

## values for index
# 'ETH' or 'USD'

url = base_url + data_url.format('2017-08-01','2017-08-02','USD')
url = 'https://api.coindesk.com/charts/data?output=json&data=close&index=USD&startdate=2014-09-01&enddate=2014-09-02&exchanges=bpi&dev=1'
```

```python
data = _getJason(url, f='text')
```

```python
data = eval(data.replace('cb(','').replace(');',''))
```

```python
data.keys()
```

```python
df = pd.DataFrame(data['bpi'], columns=['timestamp','price'])
```

```python
df['dt'] = df['timestamp'].apply(lambda x: datetime.fromtimestamp(x/1000))
```

```python
df[df['dt']>'2016-09-01'].sort_values(['dt'])
```
